```python
import matplotlib.pyplot as plt

quantiles = [0.5000, 0.7500, 0.9000, 0.9500, 0.9900, 0.9990, 0.9999]

# values obtained via scripts/t5_encoder_hf_precision_parity.py
hf_bf16 = [0.0006, 0.0011, 0.0018, 0.0022, 0.0033, 0.0047, 0.0080]
nai_bf16 = [0.0003, 0.0005, 0.0008, 0.0010, 0.0014, 0.0022, 0.0037]

hf_fp16 = [4.7763e-05, 9.2119e-05, 1.4318e-04, 1.7839e-04, 2.4898e-04, 3.5743e-04, 6.5168e-04]
nai_fp16 = [3.7434e-05, 7.2266e-05, 1.1227e-04, 1.3884e-04, 1.9671e-04, 2.7551e-04, 4.2810e-04]

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot BF16 comparison
ax1.plot(quantiles, hf_bf16, 'o-', label='HF pure-bf16', color='blue')
ax1.plot(quantiles, nai_bf16, 's-', label='NAI bf16', color='green')
ax1.set_title('HF float32 vs BF16 Implementations')
ax1.set_xlabel('Quantile')
ax1.set_ylabel('Absolute Max Difference')
ax1.set_yscale('log')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Plot FP16 comparison
ax2.plot(quantiles, hf_fp16, 'o-', label='HF fp16', color='blue')
ax2.plot(quantiles, nai_fp16, 's-', label='NAI fp16', color='green')
ax2.set_title('HF float32 vs FP16 Implementations')
ax2.set_xlabel('Quantile')
ax2.set_ylabel('Absolute Max Difference')
ax2.set_yscale('log')
ax2.grid(True, alpha=0.3)
ax2.legend()

# Adjust layout
plt.tight_layout()
plt.suptitle('T5 Encoder Precision Comparison: HF vs NAI', y=1.05, fontsize=16)
plt.show()

# Create a separate figure for combined view
plt.figure(figsize=(10, 8))
plt.plot(quantiles, hf_bf16, 'o-', label='HF pure-bf16', color='blue')
plt.plot(quantiles, nai_bf16, 's-', label='NAI bf16', color='green')
plt.plot(quantiles, hf_fp16, 'o--', label='HF fp16', color='darkblue')
plt.plot(quantiles, nai_fp16, 's--', label='NAI fp16', color='darkgreen')
# plt.title('T5 Encoder Precision Comparison (All Formats)')
plt.title('T5v1.1 XXL Encoder precision. Absmax difference from reference (HF float32), by quantile')
plt.xlabel('Quantile')
plt.ylabel('Absolute Max Difference (log scale)')
plt.yscale('log')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
